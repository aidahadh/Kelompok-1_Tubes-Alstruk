## DEFINE FUNCTION
FUNGSI PENCARIAN LINEAR UNTUK KEBUTUHAN INSERT DATA OLEH USER

```python
def linear(a, x):
    for i in range(len(a)):
        if a[i] == x:
            return True
    return False
```

## DEFINE CLASS OBJECT
CLASS HEWAN SEBAGAI INDUK DAN PENYIMPANAN DATA DARI INPUT USER

```python
class hewan:
    def __init__(self):
        self.data = {}
        self.check=[]
        
    def insert(self,jenis,nama,spesies,id_hewan,ciri):
        find = linear(self.check,jenis)
        self.check.append(jenis)
        if find==False:
            self.data[jenis]={nama:{'spesies':{spesies:{'id':id_hewan,'ciri':ciri}}}}#jenis baru
        elif find==True:
            spec=[]
            for i in self.data[jenis]:
                spec.append(i)
            find2 = linear(spec,nama)
            if find2== False:
                self.data[jenis][nama]={'spesies':{spesies:{'id':id_hewan,'ciri':ciri}}}#nama baru
            elif find2== True:
                self.data[jenis][nama]['spesies'][spesies]={'id':id_hewan,'ciri':ciri}#spesies baru
    
    def id_char(self,jenis,nama,spesies):
        print('\n■ ID',spesies,'\t:',self.data[jenis][nama]['spesies'][spesies]['id'],'\n■ Dengan ciri-ciri :\n')
        for i in self.data[jenis][nama]['spesies'][spesies]['ciri']:
            print('●',i)
        
    def showdatabase(self):
        print('■'*20,'DATABASE KEBUN BINATANG','■'*20)
        for a in self.data:
            print('■■■■',a.upper(),'■■■■')
            for b in self.data[a]:
                print(b)
                for c in self.data[a][b]:
                    for d in self.data[a][b][c]:
                        print('\t●',d)
                        for e in self.data[a][b][c][d]:
                            print('\t\t',e,'\t= ',self.data[a][b][c][d][e])
        dic=input('Ingin melihat data dengan format dictionary?  [y/n]\t')
        if dic=='y':
            print(self.data)
            
    def lookdata(self,jenis,nama,x):
        target=self.data[jenis][nama]['spesies']
        c=[]
        for i in target:
            c.append(i)
        return linear(c,x)
    
    def sounds(self):
        if self.data['mamalia']== 'paus ':
            print('suara paus')
        elif self.data['mamalia']== 'singa':
            print('suara singa')
        elif self.data['aves']== 'elang':
            print('suara elang')
        elif self.data['insekta']== 'lebah':
            print('suara lebah')
            
    def mobility(self):
        if self.data['mamalia']== 'paus ':
            print('berenang')
        elif self.data['mamalia']== 'singa':
            print('berlari')
        elif self.data['aves']== 'elang':
            print('terbang')
        elif self.data['insekta']== 'lebah':
            print('terbang')
```

## DEFINE SUB CLASS OBJECT
CLASS SINGA, PAUS, LEBAH, DAN ELANG SEBAGAI SUB-CLASS DAN PEMANGGILAN INFORMASI MENGENAI HEWAN-HEWAN TERTENTU

```python
class singa(hewan):
    def __init__(self):
        super().__init__()
    def sounds(self):
        print('Singa berkomunikasi melalui vokalisasi seperti auman keras,',
              ' dengan meninggalkan tanda aroma, dan melalui isyarat visual, seperti kegelapan surai.',
              ' Singa Jantan mengaum untuk melindungi wilayah mereka dari pejantan lain, untuk menarik pasangan kawin,',
              ' dan untuk mencegah pemangsa lainnya.\n')
    def mobility(self):
        print('Mobilitas Singa adalah dengan berlari/berjalan dengan 4 kakinya.',
              ' Singa tidak dapat berlari untuk jarak jauh tetapi hanya dalam jarak pendek.',
              ' Bagian bawah cakar singa memiliki bantalan lembut dengan cakar yang dapat ditarik di setiap jari kaki,',
              ' yang memungkinkan mereka berlari untuk jarak pendek. Tumit mereka tidak menyentuh tanah ketika berjalan.',
              ' Ini karena mereka memiliki jari kaki yang besar dan bantalan tersebut,',
              ' yang memungkinkan mereka untuk bergerak dengan tenang.\n')
    def breath(self):
        print('Singa adalah mamalia dan sewajarnya bernafas dengan paru-paru\n\n')

class paus(hewan):
    def __init__(self):
        super().__init__()
    def sounds(self):
        print('Mamalia laut sangat bergantung pada indra pendengaran untuk berkomunikasi dan makan,',
              ' karena samudra menghamburkan cahaya yang mengakibatkan Penglihatan kurang efektif.',
              ' Indra penciuman juga terbatas karena molekul berdifusi lebih lambat di air.',
              ' Paus menggunakan Vokalisasi untuk berkomunikasi. Suara yang dihasilkan oleh paus bisa',
              'mencapai hingga 188 db, tujuannya adalah untuk menjaga jarak antar individu, mengenali ',
              'spesies, bahkan untuk mencari makanan.\n')
    def mobility(self):
        print('Karena hidup di lautan, mobilitas paus bergantung kepada sirip dan ekornya.',
              ' Paus berenang dengan gerakan ekor vertikal','Menurut ahli dari University In Glassboro,',
              ' Kenneth Lacovara, alasan paus menggerakkan ekornya ke atas dan ke bawah adalah karena paus',
              'berevolusi dari mamalia sekitar 50 juta tahun yang lalu. Mamalia empat kaki yang tinggal di',
              'darat menggerakkan tulang belakangnya ke atas dan ke bawah ketika berlari.\n')
    def breath(self):
        print('Sebagaimana Mamalia, meskipun hidup di lautan, Paus bernafas dengan paru-paru\n\n')

class elang(hewan):
    def __init__(self):
        super().__init__()
    def sounds(self):
        print('Suara dari Elang adalah kombinasi dari siulan dan "piping" bernada tinggi.',
              ' Jeritan Elang atau suara memekiknya sering diberikan sebagai tanggapan terhadap manusia yang',
              'mendekati Elang dan merupakan tangisan bernada tinggi yang memiliki tiga hingga lima nada yang ',
              'terdengar seperti burung camar diikuti oleh enam atau tujuh nada cepat.\n')
    def mobility(self):
        print('Karena hidup di darat dan udara(dominan), cara mobilitas Elang adalah terbang.',
              'Elang terbang dengan cara mengepakan sayapnya. Saat sayap burung dikepakan, udara akan didorong kebawah.',
              ' Dorongan kebawah akan menghasilkan gaya yang berlawanan sehingga mengangkat tubuh burung keatas.',
              ' Untuk mengendalikan gerakan baik menukik, membelok, burung menggunakan ekornya.\n')
    def breath(self):
        print('Elang bernafas dengan paru-paru. Namun Ketika mengudara,',
              ' Elang menggunakan udara kantong/pundi- pundi udara sebagai alat pernapasannya.',
              ' ketika  tidak terbang, maka Elang mengisi pundi-pundi udaranya.\n')
        
class lebah(hewan):
    def __init__(self):
        super().__init__()
    def sounds(self):
        print('Manusia mendengar lebah mendengung, namun sebenarnya suara tersebut dihasilkan oleh',
              'kepakan sayap serangga yang cepat saat mereka terbang di udara dan menciptakan getaran',
              'yang dideteksi telinga manusia sebagai dengung.\n\tBzzzzzzz\n')
    def mobility(self):
        print('lebih kurang sama seperti Elang, mobilitas lebah dominan terbang. ',
              'Sayap mereka tidak kaku, tetapi *twist-and-rotate* saat terbang.',
              ' Sayap lebah membuat gerakan menyapu pendek dan cepat ke depan dan belakang, depan dan belakang.',
              ' Gerakan ini menciptakan daya angkat yang cukup untuk memungkinkan lebah terbang.\n')
    def breath(self):
        print('Lebah juga memiliki sistem peredaran darah terbuka, yang berarti bahwa mereka tidak memiliki',
              'pembuluh darah atau arteri, melainkan semua organ internal mereka bermandikan cairan yang disebut',
              '*hemolymph* (campuran darah dan cairan limfatik). Lebah bernafas melalui struktur kompleks jaringan',
              'trakea dan kantung udara.\n\n')
```

## MAIN PROGRAM
KAMI MENGAPLIKASIKAN POLYMORPH DALAM OOP SEBAGAI PROGRAM YANG DIGUNAKAN UNTUK MENAMBAHKAN DATA MAUPUN MENCARI INFORMASI SEPUTAR HEWAN DI KEBUN BINATANG

```python
hewan=hewan()
while True:
    print('\n','■'*50,'\n   SELAMAT DATANG DI KEBUN BINATANG~\n','■'*50)
    user=input('\n1. Login Petugas\n2. Pengunjung\n3. Stop Program\n\t *[1/2/3]')
    if user=='1':
        loop=True
        while loop== True:
            uname=input('>\tMasukkan Username\t:')
            pword=input('>\tMasukkan Password\t:')
            if uname=='penjagamonyet' and pword=='pendek':
                print('■'*40,'>>>\nHALO KAKAK PETUGAS!')
                while True:
                    update=input('\n>\t APA YANG INGIN KAKAK LAKUKAN ?\n1.Update Database\n2.Lihat Database\n3.Tidak Ada\n\t')
                    if update=='1':
                        for i in range(int(input('Berapa jumlah data yang ingin dimasukkan? :\n\t'))):
                            print('='*50,'Data ke',i+1,'\n')
                            jenis = input('●Masukkan nama jenis dari hewan [mamalia/aves/insekta] :\n\t')
                            id_hewan = int(input('●Masukkan id hewan [numerik] :\n\t'))
                            spesies = input('●Masukkan nama spesies dari hewan [ex : singa afrika; elang jawa] :\n\t')
                            nama = spesies[:5]
                            ciri=[]
                            for j in range(int(input('●Berapa ciri-ciri yang ingin dimasukkan? :\t'))):
                                print(j+1,end='.)')
                                ciri.append(input('\t'))
                            hewan.insert(jenis,nama,spesies,id_hewan,ciri)
                            print('\n\t=========Data ke',i+1,'berhasil ditambahkan!=========\n')
                    elif update=='2':
                        hewan.showdatabase()
                    else:
                        break
                loop=False
            elif uname!='penjagamonyet' and pword!='pendek':
                print('\nxxx=======username atau password belum sesuai, silahkan coba lagi=======xxx\n')
    elif user=='2':
        print('■'*40,'>>>\nHALO KAKAK PENGUNJUNG!\n')
        search=input('>\t HEWAN APA YANG INGIN KAKAK CARI TAHU? [ex : singa afrika; elang jawa] :\n\t')
        if search[:5]=='singa':
            if hewan.lookdata('mamalia',search[:5],search)==True:
                print('\n\t\t■■',search.upper())
                singa=singa()
                hewan.id_char('mamalia',search[:5],search)
                print('\t\t DESKRIPSI\n')
                singa.mobility()
                singa.sounds()
                singa.breath()
            elif hewan.lookdata('mamalia',search[:5],search)==False:
                print('Spesies tidak ditemukan')
        elif search[:5]=='paus ':
            if hewan.lookdata('mamalia',search[:5],search)==True:
                print('\n\t\t■■',search.upper())
                paus=paus()
                hewan.id_char('mamalia',search[:5],search)
                print('\t\t DESKRIPSI\n')
                paus.mobility()
                paus.sounds()
                paus.breath()
            elif hewan.lookdata('mamalia',search[:5],search)==False:
                print('Spesies tidak ditemukan')
        elif search[:5]=='elang':
            if hewan.lookdata('aves',search[:5],search)==True:
                print('\n\t\t■■',search.upper())
                elang=elang()
                hewan.id_char('aves',search[:5],search)
                print('\t\t DESKRIPSI\n')
                elang.mobility()
                elang.sounds()
                elang.breath()
            elif hewan.lookdata('aves',search[:5],search)==False:
                print('Spesies tidak ditemukan')
        elif search[:5]=='lebah':
            if hewan.lookdata('insekta',search[:5],search)==True:
                lebah=lebah()
                print('\n\t\t■■',search.upper())
                hewan.id_char('insekta',search[:5],search)
                print('\t\t DESKRIPSI\n')
                lebah.mobility()
                lebah.sounds()
                lebah.breath()
            elif hewan.lookdata('insekta',search[:5],search)==False:
                print('Spesies tidak ditemukan')
        else:
            print('maaf hewan tidak ditemukan')
            
            
    elif user=='3':
        break
print('\n\n','=■'*12,'TERIMA KASIH TELAH MENGGUNAKAN PROGRAM INI, SAMPAI JUMPA!','■='*12,'\n')
```
